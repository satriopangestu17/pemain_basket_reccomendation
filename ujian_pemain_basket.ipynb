```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv('all_seasons.csv')

df.columns.tolist()
```

```python
dfbasket = df[['player_name','age','player_height','player_weight','pts','reb']]
dfbasket


dftarget = dfbasket[dfbasket['age'] <= 25][dfbasket['player_height'] >= 180][dfbasket['player_weight'] <= 90][dfbasket['pts'] >= 6][dfbasket['reb'] >= 3]
dftarget
# [['player_name','player_height','player_weight','pts','reb']]





labeltarget = 'Potential for Recruitment (Target)'
dftarget['label'] = labeltarget
dftarget

```

```python
# standard scaller
from sklearn.preprocessing import StandardScaler

## agar aman standardisasi datas
stdscaler = StandardScaler()

##target tidak diikutsertakan
datatarstd = stdscaler.fit_transform(dftarget.iloc[:,1:6])  ## exclude target
stdscaler.fit_transform(dftarget.iloc[:,1:6]).shape
# dftarget.iloc[:,1:6]
datatarstd
```

```python
dftargetn = pd.DataFrame(
    datatarstd,
    columns=['age', 'player_height', 'player_weight', 'pts', 'reb']
)
dftargetn

labeltarget = 'Potential for Recruitment (Target)'
dftargetn['label'] = labeltarget
dftargetn
dftargetn['player_name'] = dfbasket['player_name']
dftargetn
```

```python
dfnontarget = dfbasket.drop(dftarget.index)
dfnontarget

labelnontarget = 'Not for Recruitment (Non Target)'
dfnontarget['label'] = labelnontarget
dfnontarget
```

```python
# standard scaller
from sklearn.preprocessing import StandardScaler

## agar aman standardisasi datas
stdscaler = StandardScaler()

##target tidak diikutsertakan
datanontarstd = stdscaler.fit_transform(dfnontarget.iloc[:,1:6])  ## exclude target
stdscaler.fit_transform(dfnontarget.iloc[:,1:6]).shape
# dftarget.iloc[:,1:6]
datanontarstd
```

```python
dfnontargetn = pd.DataFrame(
    datanontarstd,
    columns=['age', 'player_height', 'player_weight', 'pts', 'reb']
)
dfnontargetn

labelnontarget = 'Not for Recruitment (Non Target)'
dfnontargetn['label'] = labelnontarget
dfnontargetn
dfnontargetn['player_name'] = dfbasket['player_name']
dfnontargetn
```

```python
dfcomplete = dftargetn.append(dfnontargetn, ignore_index= True) 
dfcomplete
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    dfcomplete[['age', 'player_height', 'player_weight', 'pts', 'reb']],
    dfcomplete['label'],
    test_size = .18, random_state=10
)
```

```python
k = StratifiedKFold(n_splits = 100, random_state=None, shuffle=False)
k
```

```python
datax = dfcomplete[['age', 'player_height', 'player_weight', 'pts', 'reb']].values
datax
```

```python
def k_value():  ## sama saja bukan fungsi kalau tidak di isi ()-nya
    k = round((len(x_train)+len(x_test)) ** .5)
    if (k % 2 == 0):
        return k + 1    ##return have to be in function
    else:
        return k
k_value()

```

```python
print('LR Score:',round(cross_val_score(LogisticRegression(solver = 'lbfgs', multi_class = 'auto'),x_train,y_train, cv = 5).mean() * 100), 'percent')
print('DTC Score:',round(cross_val_score(DecisionTreeClassifier(),x_train,y_train, cv = 5).mean() * 100), 'percent')
print('KNN Score:',round(cross_val_score(KNeighborsClassifier(n_neighbors = k_value()),x_train,y_train, cv = 5).mean() * 100), 'percent')
```

```python
##TUNING
# LogisticRegression parameter yang akan dituned + nilai yang mungkin
penalty = ['l1', 'l2', 'elasticnet', 'none']
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
max_iter = [1, 10, 100, 1000, 10000]

param = {'penalty': penalty, 'solver': solver, 'max_iter': max_iter}

param
```

```python
from sklearn.model_selection import RandomizedSearchCV

model = LogisticRegression()
#randomized search

modelrs = RandomizedSearchCV(estimator=model, param_distributions = param, cv=5)  ## 5 fold
```

```python
modelrs.fit(x_train, y_train)
modelrs.best_params_
```

```python
modellr = LogisticRegression(solver='liblinear',max_iter=1000)
modellr.fit(x_train, y_train)
modellr.score(x_train, y_train)
```

```python
##TUNING
# Decision tree parameter yang akan dituned + nilai yang mungkin
criterion = ['gini', 'entropy']
splitter = ['best', 'random']
# max_features = [1, 10, 100, 1000, 10000]

paramtree = {'criterion': criterion, 'splitter': splitter}

paramtree
```

```python
from sklearn import tree
modeltree = tree.DecisionTreeClassifier() 
modeltree.fit(x_train, y_train)
modeltree.score(x_train, y_train)
```

```python
#### KNeighborsClassifier  ###
from sklearn.neighbors import KNeighborsClassifier
modelkn = KNeighborsClassifier(n_neighbors = k_value())
modelkn.fit(x_train, y_train)
modelkn.score(x_train, y_train)
```
