```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_breast_cancer
bc = load_breast_cancer()
dir(bc)
```

```python
df = pd.DataFrame(
    bc['data'], columns= bc['feature_names']
)
df.head()
df['target'] = bc['target']
bc['target_names']
```

```python
df.isnull().sum()
```

```python
i = np.array(df.corr()['target'][df.corr()['target']< -0.5].index)
dfnew = df[i]
dfnew['target'] = df['target']
dfnew
```

## Standardisasi

```python
#karena ada data yang tinggi / rendah sendiri
```

```python
std = StandardScaler()
datastd = std.fit_transform(dfnew[i])
#target ga ikut di standardisasi
datastd
```

```python
#check outlier, -2.5 < z-score > 2.5
dfzscore = pd.DataFrame(datastd)
dfzscore

#masukin target setelah standardisasi
dfzscore['target'] = df['target']
dfzscore
```

```python
dfzscore[(dfzscore < -2.5) | (dfzscore > 2.5)] = np.NaN
dfzscore = dfzscore.dropna()
dfzscore
```

```python
np.array(dfzscore)
```

```python
datastd = np.array(dfzscore.iloc[:,:-1])
```

## PCA

```python
# Pricipal Component Analyze = untuk memperdikit feature. dataframe ini ada 15 feature.
```

```python
pca = decomposition.PCA(n_components=3)
datapca = pca.fit_transform(datastd)
datapca
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
#feature pertama 74%, dst

#inverse ke value semula
#inverse PCA dulu baru inverse standard scaler
std.inverse_transform(pca.inverse_transform(datapca[0]))
```

# Pipeline = Standardize + PCA

```python
pipe = make_pipeline(
    StandardScaler(),
    decomposition.PCA(n_components = 3)
)
stdpca =  pipe.fit_transform(dfnew.iloc[:,:-1])
stdpca[0]
```

```python
#inverse ke value semula
pipe.inverse_transform(stdpca)[0]
```

# Splitting Dataset

```python
x = datapca
y = np.array(dfzscore['target'])
y.shape
```

```python
xtr, xts, ytr, yts = train_test_split(
    x, y, test_size = 0.15
)
print(xtr.shape)
print(xts.shape)
print(ytr.shape)
print(yts.shape)
```

## K-Nearest Neighbors

```python
k = round(xtr.shape[0] ** 0.5)
k
```

```python
model = KNeighborsClassifier(n_neighbors = k)
model.fit(xtr,ytr)
```
